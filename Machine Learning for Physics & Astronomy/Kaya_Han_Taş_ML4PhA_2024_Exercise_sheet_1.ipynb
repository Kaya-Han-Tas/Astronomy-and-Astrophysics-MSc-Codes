### Answer:

### a) Which possibility is the most likely, a priori, given the assumptions above: two cats, two dogs, or one cat and one dog?

* The probability of owning a Cat or Dog are as follows.

$$ P(Cat) = 0.5 $$
$$ P(Dog) = 0.5 $$

* We should note that **the probabilities of all combinations should add up to 1!**
* We use the following formula since the events are **Mutually Exclusive and Exhaustive**!

$$ P(A,B) = P(A) * P(B) $$

* From here we get the following result.

| A | B | P(A,B) |
| :-: | :-: | :------: |
| Cat | Cat | 0.25 |
| Dog | Dog | 0.25 |
| Cat | Dog | 0.25 |
| Dog | Cat | 0.25 |.99 |

### b) Suppose I ask my neighbor if he has any dogs, and he says yes. What is the probability that one pet is a cat?

* The chance of owning a Cat and a Dog is mutually exclusive.

$$ P(Cat) = 0.5 $$
$$ P(Dog) = 0.5 $$

* We can create a table for possible outcomes.

| A | B | P(A,B) |
| :-: | :-: | :------: |
| Cat | Cat | 0.25 |
| Dog | Dog | 0.25 |
| Cat | Dog | 0.25 |
| Dog | Cat | 0.25 |

* We know that the neighbor owns a Dog.
* Which means that either A or B is Dog.
* **That eliminates the first row!**

| A | B | P(A,B) |
| :-: | :-: | :------: |
| Dog | Dog | 0.25 |
| Cat | Dog | 0.25 |
| Dog | Cat | 0.25 |

* From the table, the probability of owning a Dog is $P(Dog) = 0.75$.
* We can **normalize** the probabilities using this as follows:

$$ \frac{P(A,B)}{P(Dog)} $$

* From here we get the Normalized Table of the following.

| A | B | P(A,B) |
| :-: | :-: | :------: |
| Dog | Dog | 0.33 |
| Cat | Dog | 0.33 |
| Dog | Cat | 0.33 |

* From here, the Probability of the other pet being a cat is **the probability of the 2nd and 3rd row added up!**

$$P(Dog, Cat) = 0.33 + 0.33 = 0.66 = \frac{2}{3}$$

### 1.2 Monty Hall Problem (2 points)

*Topics: joined and marginal distributions, probability tables*

On a game show, a contestant is told the following rules:


> There are three doors, labelled 1, 2, 3. A single prize has been hidden behind one of them. You get to select one door. Initially your chosen door will not be opened. Instead, the gameshow host will open one of the other two doors, and *he will do so in such a way as not to reveal the prize.* For example, if you first choose door 1, he will then open one of doors 2 and 3, and it is guaranteed that he will choose which one to open so that the prize will not be revealed. At this point, you will be given a fresh choice of door: you can either stick with your first choice, or you can switch to the other closed door. All the doors will then be opened and you will receive whatever is behind your final choice of door.

Imagine that the contestant chooses door 1 first; then the gameshow host opens door 3, revealing nothing
behind the door, as promised. Should the contestant (a) stick with door 1, or (b) switch to door 2, or (c) does it make no difference? You may assume that initially, the prize is equally likely to be behind any of the 3 doors.

Use probability tables to justify your answer.

### Answer:

* First, let's create the Probability table for each possible outcomes.
* **In the Table:**
    * **Prize Door:** The door that has the prize behind it.
    * **Chosen Door:** The door that the contestant chooses initially.
    * **Opened Door:** The door that the host opens.
    * **Result (No Switch):** If the contestant does not change his chosen door after the host opens one door.
    * **Result (Switch):** If the contestant does change his chosen door after the host opens one door.

| Prize Door | Chosen Door | Opened Door | Result (No switch) | Result (Switch)|
| :-: | :-: | :-: | :-: | :-: |
| 1 | 1 | 2 or 3 | Win | Loss |
| 1 | 2 | 3 | Loss | Win |
| 1 | 3 | 2 | Loss | Win |
| 2 | 1 | 3 | Loss | Win |
| 2 | 2 | 1 or 3 | Win | Loss |
| 2 | 3 | 1 | Loss | Win |
| 3 | 1 | 2 | Loss | Win |
| 3 | 2 | 1 | Loss | Win |
| 3 | 3 | 1 or 2 | Win | Loss |

* Door 1 is chosen so the probability table becomes the following.

| Prize Door | Chosen Door | Opened Door | Result (No switch) | Result (Switch)|
| :-: | :-: | :-: | :-: | :-: |
| 1 | 1 | 2 or 3 | Win | Loss |
| 2 | 1 | 3 | Loss | Win |
| 3 | 1 | 2 | Loss | Win |

* **The chance of winning if the contestant does not switch is 1/3!**
* **The chance of winning if the contestant does switch is 2/3!**
* The host opens the Door 3 which leaves us with the following Probability table.

| Prize Door | Chosen Door | Opened Door | Result (No switch) | Result (Switch)|
| :-: | :-: | :-: | :-: | :-: |
| 1 | 1 | 2 or 3 | Win | Loss |
| 2 | 1 | 3 | Loss | Win |

### a) Should the contestant stick with door 1
### b) Should the contestant switch to door 2
### c) Does it make no difference?

* In here, we might say there is a 50/50 chance to get the price whether we switch or not but that isn't the case!
* **The probabilities should add up to 1!**
* For that reason, since the probability of the prize being behind door 1 was 1/3 **and it stays the same even when one of the doors is revealed**, **the probability of the Door 2 having the price is 2/3!**

$$ P(Door1) = 1/3 $$
$$ P(Door2) = 2/3 $$

* **In other words:**
    * **Before Door Reveal:** The probability of chosen door having the prize is 1/3, while the probability of the unchosen doors having the prize is also 1/3.
    * **After Door Reveal:** The probability of the chosen door having the prize still is 1/3, **while the probability of the unchosen doors having the prize is now 2/3!**

* **So the Contestant should switch to Door 2 since it has a better chance of containing the prize!**

### 1.4 Sampling from a Probabilistic Model (2 points)


*Topic: frequentist probabilities*

Let's say we have a probabilistic model with

| X | p(X) |
|------|------|
| Sun (0)   | 0.2 |
| Rain (1)  | 0.3 |
| Snow (2)  | 0.5 |

a. Write a sampler that randomly samples from $p(X)$ values X = 0, 1, 2. In your solution, you are allowed to use the function `numpy.random.random()` but other than this, do not use any other imported functions.

### Answer

### a)

```python
#We import the libraries.
from numpy.random import random
import numpy as np
import matplotlib.pyplot as plt

#We define a function that randomly samples a value from [0,1,2] using the Probability values.
def return_random_sample(sample_no = 1):
    """Takes the number of samples that the user wants to generate. Returns the
    randomly generated numbers from the list [0,1,2] depending on the probabilities
    of each number.
    
    Input:
    ----------------------------------------------------------------------------
    sample_no: Number of Samples/Random Numbers that will be generated.
    
    Output:
    ----------------------------------------------------------------------------
    sample_list: List containing the random numbers generated.
    
    """
    #We define the intervals.
    #Sun => Between 0.0 and 0.2
    #Rain => Between 0.2 and 0.5 (You add up 0.3 to the previous interval's upper limit)
    #Snow => Between 0.5 and 1 (You add up 0.5 to the previous interval's upper limit)
    prob_intervals = [[0.0, 0.2], [0.2, 0.5], [0.5, 1.0]]

    #We get a random number using the np.random.random() function.
    #Note that it samples a float value between 0 and 1.
    random_number = np.random.random(sample_no)

    #Now we define a list that will store the randomly generated numbers.
    sample_list = []
    
    #Now we want to check this number, find which interval it ends up on and then sample either 0, 1 or 2 as
    #follows.
    #We use enumerate to use our Probability intervals and use their indices.
        #We check which interval the random_number we generated ends up on by using a for loop over the
        #probability interval (prob_intervals) we have defined previously.
        #Then we use the index of that interval to get the number/sample from [0,1,2].

    for generated_number in random_number:
        for index_interval, (lower_lim, upper_lim) in enumerate(prob_intervals):
            if lower_lim <= generated_number < upper_lim:
                sample_list.append(index_interval)

    #We return the samples.
    return sample_list

#We use the function to sample 50 data points from the given table and probabilities.
sample_list_result = return_random_sample(sample_no = 50)
print("The Generated Samples:", "\n", 50*"-", "\n", sample_list_result)
```

```python
#We first define a list containing the number of samples we will generate.
sample_no_list = [5, 10, 20, 50, 100, 200, 500, 1000]

#Since we have defined 8 number of samples, we'll create a figure that has plots with 4x2 dimension.
figure, axes = plt.subplots(2, 4, figsize = (15, 10))
axes = axes.ravel()

#We now define a loop that will go over this list and generate histograms for different number of samples.
for i, sample_no in enumerate(sample_no_list):
    #We get our samples from the function we have defined.
    samples = return_random_sample(sample_no = sample_no)

    #We then plot the Histogram for the given sample no.
    #Note that we substract 0.5 to center the bins.
    bins = np.arange(4) - 0.5
    axes[i].hist(samples, bins = bins, density = True, edgecolor = "black")

    #We set the Axes, Title and Labels of the histogram.
    axes[i].set_title(f"Number of Samples: N = {sample_no}")
    axes[i].set_xlabel("X value")
    axes[i].set_ylabel("P(X) value")
    axes[i].set_xlim(-1, 3)

    #We define the ticks to show the number and also what it represents.
    axes[i].set_xticks([0, 1, 2])
    axes[i].set_xticklabels(["Sun\n(0)", "Rain\n(1)", "Snow\n(2)"])

    #We use plt.tight_layout() to prevent plots clipping through one another.
    plt.tight_layout()
```

### 1.5 Conditional Probabilities (3 points)

*Topics: frequentist probabilities, conditional probabilities*

Consider the probability table:

| X | Y | p(X,Y) |
|------|------| ----- |
| Hot (0)   | Sun (0) | 0.4 |
| Hot (0) | Rain (1) | 0.1 |
| Cold (1)  | Sun (0) | 0.2 |
| Cold (1)  | Rain (1) | 0.3 |



a. Implement a sampling algorithm that samples $(X, Y)$ pairs.

### Answer:

```python
#We define the function similar to the previous one.
def return_random_sample_joint(sample_no):
    """Takes the number of samples that the user wants to generate. Returns the
    randomly generated numbers for X and Y from the list [0,1] depending on the 
    probabilities of each number.
    
    Input:
    ----------------------------------------------------------------------------
    sample_no: Number of Samples/Random Numbers that will be generated.
    
    Output:
    ----------------------------------------------------------------------------
    sample_list: List containing the random numbers generated in [X,Y] form.
    
    """
    #We define the intervals.
    #Hot & Sun => Between 0.0 and 0.4
    #Hot & Rain => Between 0.4 and 0.5
    #Cold & Sun => Between 0.5 and 0.7
    #Cold & Rain => Between 0.7 and 1.0
    prob_intervals = [[0.0, 0.4], [0.4, 0.5], [0.5, 0.7], [0.7, 1.0]]

    #We get a random number using the np.random.random() function.
    #Note that it samples a float value between 0 and 1.
    random_number = np.random.random(sample_no)

    #Now we define a list that will store the randomly generated numbers.
    sample_list = []
    
    #Now we want to check this number, find which interval it ends up on and then sample either 0, 1 or 2 as
    #follows.
    #We use enumerate to use our Probability intervals and use their indices.
        #We check which interval the random_number we generated ends up on by using a for loop over the
        #probability interval (prob_intervals) we have defined previously.
        #Then we use the index of that interval to get the number/sample from [0,1,2].

    for generated_number in random_number:
        for index_interval, (lower_lim, upper_lim) in enumerate(prob_intervals):
            if lower_lim <= generated_number < upper_lim:
                #Depending on index of the interval we are on we define our X,Y Values.
                #Simpler way to do this: sample_list.append([index_interval // 2, index_interval % 2])
                if index_interval == 0:
                    sample_list.append([0,0])

                elif index_interval == 1:
                    sample_list.append([0,1])

                elif index_interval == 2:
                    sample_list.append([1,0])

                elif index_interval == 3:
                    sample_list.append([1,1])

    #We return the samples.
    return sample_list

#We use the function to sample 50 data points from the given table and probabilities.
sample_list_result = return_random_sample_joint(sample_no = 50)
print("The Generated Samples:", "\n", 50*"-", "\n", sample_list_result)
```

Mathematically, the PDF of a sum of two random variables is the convolution of their individual PDFs (do you understand why?). Let's call their PDFs $f(x)$ and $g(x)$, then the convolution is defined as

$$ (f \circledast g)(x) = \int_{-\infty}^{\infty} f(\tau) g(x - \tau) d\tau \, .$$

Program this convolution yourself for the two normal distributions $\mathcal{N}(x|\mu=2, \sigma=1)$ and $\mathcal{N}(x|\mu=4, \sigma=2)$ and show that you can recover the same connections seen already for the sets $\vec{x}$, $\vec{y}$ and $\vec{z}$.

**Hint**: You are **not** allowed to use a predefined convolution function. Approximate the integral by a sum over small elements with width $\Delta \tau$.

### Answer:

* This time we won't be using np.random.normal() since now we will have our x values.
* We will instead use the formula below.

$$N(x|\mu,\sigma)=\frac{1}{\sqrt{2\pi\sigma^{2}}}e^{-\frac{1}{2}\frac{(x-\mu)^{2}}{\sigma^{2}}}$$

```python
#Now we define two separate Normal Distributions for f(x) and g(x) by using the formula above as follows.

#f(x) => Mu = 2 and Sigma = 1
def f(x):
    """Takes a list of x values and returns a Normal Distribution with Mean of 2
    and Standard Deviation of 1.
    
    Input:
    ----------------------------------------------------------------------------
    x: List that contains the x values.
    
    Output:
    ----------------------------------------------------------------------------
    result: List containing the y values of the Normal Distribution f(x).
    
    """
    result_1 = 1 / np.sqrt(2 * np.pi * pow(1, 2))
    result_2 = np.exp(-(1/2) * (pow(x-2, 2)/pow(1, 2)))
    result = result_1 * result_2

    return result

#g(x) => Mu = 4 and Sigma = 2
def g(x):
    """Takes a list of x values and returns a Normal Distribution with Mean of 4
    and Standard Deviation of 2.
    
    Input:
    ----------------------------------------------------------------------------
    x: List that contains the x values.
    
    Output:
    ----------------------------------------------------------------------------
    result: List containing the y values of the Normal Distribution g(x).
    
    """
    result_1 = 1 / np.sqrt(2 * np.pi * pow(2, 2))
    result_2 = np.exp(-(1/2) * (pow(x-4, 2)/pow(2, 2)))
    result = result_1 * result_2

    return result
    
#We also define a range for x values we will look into.
#We'll use the bin spacing we used in the previous parts for the upper and lower limits.
x_range = np.linspace(-2, 14, 10000)

#Now we define a function for the convolution.
def convolution(f, g, x_range):
    """Takes f(x) and g(x) functions alongside the range of x values defined.
    Returns the convolution of f(x) and g(x).
    
    Input:
    ----------------------------------------------------------------------------
    f: Function that returns a Normal Distribution with Mean of 2 and Standard
    Deviation of 1. (y values)
    g: Function that returns a Normal Distribution with Mean of 4 and Standard
    Deviation of 2. (y values)
    x_range: List that contains the x values that will go into the f and g
    functions.
    
    Output:
    ----------------------------------------------------------------------------
    convolution_list: List containing the y values of the convolution of f(x)
    and g(x).
    
    """
    #We define a list to store the convolution results.
    convolution_list = np.zeros_like(x_range)

    #Now using the x_range we have defined, we get delta tau i.e. the steps.
    delta_tau = x_range[1] - x_range[0]

    #Now we define a loop to do the convolution as follows.
    #Note that this uses the formula given in the question.
    for i, x in enumerate(x_range):
        convolution_list[i] = np.sum(f(x_range) * g(x - x_range) * delta_tau)

    #Now we normalize all of the convolution results as follows.
    convolution_list /= np.sum(convolution_list * delta_tau)

    return convolution_list

#Now we use our function to get the convolution of the two datasets as follows.
convolution_result = convolution(f, g, x_range)

#We plot our results.
plt.plot(x_range, f(x_range), label = "$f (x)$")
plt.plot(x_range, g(x_range), label = "$g (x)$")
plt.plot(x_range, convolution_result, label = "Convolution $(f \circledast g)(x)$")
plt.xlabel("X")
plt.ylabel("P(X)")
plt.title("Convolution of f(x) and g(x)")
plt.legend()
plt.grid(True)
plt.show()
```

$$\mu = \Sigma{xf(x)\Delta\tau}$$
$$\sigma = \Sigma{(x-\mu)^{2}f(x)\Delta\tau}$$
$$\sigma^{2} = (\Sigma{(x-\mu)^{2}f(x)\Delta\tau})^{2}$$


```python
#We first define Delta Tau as follows. (Same as the one in the convolution function)
delta_tau = x_range[1] - x_range[0]

#Now we calculate the Means, Standard Deviations and Variances of all the distributions we have.
mean_f = np.sum(x_range * f(x_range)) * delta_tau
mean_g = np.sum(x_range * g(x_range)) * delta_tau
mean_convolution = np.sum(x_range * convolution(f, g, x_range)) * delta_tau

std_f = np.sqrt(np.sum((x_range - mean_f)**2 * f(x_range)) * delta_tau)
std_g = np.sqrt(np.sum((x_range - mean_g)**2 * g(x_range)) * delta_tau)
std_convolution = np.sqrt(np.sum((x_range - mean_convolution)**2 * convolution(f, g, x_range)) * delta_tau)

var_f = std_f**2
var_g = std_g**2
var_convolution = std_convolution**2

#We print our results.
print(90 * "-")
print("f(x): ")
print("Mean: ", mean_f)
print("Standard Deviation: ", std_f)
print("Variance: ", var_f)
print(90 * "-")
print("g(x): ")
print("Mean: ", mean_g)
print("Standard Deviation: ", std_g)
print("Variance: ", var_g)
print(90 * "-")
print("Convolution: ")
print("Mean: ", mean_convolution)
print("Standard Deviation: ", std_convolution)
print("Variance: ", var_convolution)
print(90 * "-")
```

where $\mathbf{x}$ and $\boldsymbol{\mu}$ are vectors with $D$ entries and $\mathbf{\Sigma}$ is a $D \times D$ covariance matrix. $|\mathbf{\Sigma}|$ is the determinant of this matrix.

#### a (2 points)
Consider the case where $D = 2$. Use the following three sets of ($\boldsymbol{\mu}$, $\mathbf{\Sigma}$) to create 2D contour- and 3D surface plots of the corresponding PDFs (with arbitrary contour levels)

$$ \boldsymbol{\mu_2} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_2} = \begin{pmatrix} 1 & -0.5 \\ -0.5 & 1 \end{pmatrix} \, ,$$

Implement the multivariate Gaussian yourself using numpy matrix multiplication.

### Answer:

```python
#Alongside numpy and matplotlib, we call Axes 3D for 3D Surface Plots!
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#We first define our matrices by using the matrices above as follows.
mu_list = [np.array([5, 5]),
           np.array([5, 5]),
           np.array([5, 5])]

sigma_list = [np.array([[1, 0], [0, 1]]),
              np.array([[1, -0.5], [-0.5, 1]]),
              np.array([[1, 0.7], [0.7, 1]])]

#We also define our x and y values.
#For that we create the following numpy arrays.
x_val = np.linspace(0, 10, 300)
y_val = np.linspace(0, 10, 300)

#We create a meshgrid using these two numpy arrays as follows.
#Note that this is needed to plot contour and surface plots.
#We do this to get 2 dimensional arrays that contains elements that represent x and y coordinates of a point 
#on the grid separately.
    #For example lets say x = [0,1] and y = [0,1] => X = [[0,1], [0,1]] and Y = [[0,0], [1,1]]
X_val, Y_val = np.meshgrid(x_val, y_val)

#Now we define a function that will calculate the Multivariate Gaussian as follows.
def mvgauss(x, mu, sigma):
    """Takes x values, mu matrices and sigma matrices. Calculates and Returns the 
    PDF value for Multivariate Gaussian.
    
    Input:
    ----------------------------------------------------------------------------
    x: Numpy Array containing x (Input) vector's values.
    mu: List containing matrices of Mu (Mean) vector's values as Numpy arrays.
    sigma: List containing matrices of Sigma Covariance Matrix values as Numpy 
    arrays.
    
    Output:
    ----------------------------------------------------------------------------
    result: The PDF Value for given x and mu vectors, with sigma covariance
    matrix.
    
    """
    #We define D as the length of a mu matrix since it has dimensions of D x 1!
    D = len(mu)

    #We then define the determinant of Sigma matrix since we need it for calculation.
    det_sigma = np.linalg.det(sigma)

    #We then define the inverse of the Sigma Matrix since again we need it for calculation.
    inverse_sigma = np.linalg.inv(sigma)

    #We also need the Transpose of the difference of Matrices x and mu which we can calculate as follows.
    transpose_x_mu = (x - mu).T

    #We can now calculate the Multivariate Gaussian Function's value for given x i.e. input matrix.
    #For this we use the formula given in the question above.
    #Note that the vector calculations NEED to be done with Dot Product which is why we use np.dot!
    #We can also use @ instead of np.dot, which gives us "the right vector operation".
    result_1 = 1 / np.sqrt(pow((2 * np.pi), D) * det_sigma)
    result_2 = np.exp((-1/2) * np.dot(transpose_x_mu, np.dot(inverse_sigma, (x - mu))))
    result = result_1 * result_2

    return result

#Now we can calculate the PDF values for each of our matrices as follows.
#Note: Help from AI has been used on these lines.
#The Line does the following:
    #"[mvgauss(np.array([X_val[i, j], Y_val[i, j]]), mu_list[0], sigma_list[0]) for j in range(len(X_val))]":It
#iterates over the indices of X_val for the mvgauss function we have defined.
    #"[...] for i in range(len(X_val))": Creates a nested list that iterates over X_val array to create a 2D
#array.
    #"np.array([...])": Turns the expression into a numpy array.

#Basically what it does is it iterates over all of the columns per given row. We also iterate over all the rows
#so that means we cover all of the matrix by the following lines.
#That way we calculate the Multivariate Gaussian for all rows and columns of X and Y matrices.
PDF_1 = np.array([[mvgauss(np.array([X_val[i, j], Y_val[i, j]]), mu_list[0], sigma_list[0]) for j in range(len(X_val))] for i in range(len(X_val))])
PDF_2 = np.array([[mvgauss(np.array([X_val[i, j], Y_val[i, j]]), mu_list[1], sigma_list[1]) for j in range(len(X_val))] for i in range(len(X_val))])
PDF_3 = np.array([[mvgauss(np.array([X_val[i, j], Y_val[i, j]]), mu_list[2], sigma_list[2]) for j in range(len(X_val))] for i in range(len(X_val))])

#Now we can plot our results.
#We first plot our 2D Contour Plots as follows.
#We create our figure.
figure1, axes1 = plt.subplots(1, 3, figsize = (15, 5))

#We create our Contour Plot for all PDFs.
plt.subplot(131)
plt.contour(X_val, Y_val, PDF_1)
plt.title("2D Contour Plot for $\mu_1$ and $\Sigma_1$")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(132)
plt.contour(X_val, Y_val, PDF_2)
plt.title("2D Contour Plot for $\mu_2$ and $\Sigma_2$")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(133)
plt.contour(X_val, Y_val, PDF_3)
plt.title("2D Contour Plot for $\mu_3$ and $\Sigma_3$")
plt.xlabel("X")
plt.ylabel("Y")

#We use plt.tight_layout() to prevent plots clipping through one another.
plt.tight_layout()
plt.show()
```

### 2.3 Inverse problem with Poisson distribution (2 points)

*Topic: Bayes theorem, uniform and Poisson distributions, discrete and continuous random variables*

The number of steps that a robot takes in a given amount of time is $s$, where $s$ is Poisson distributed with mean value $\lambda$. Derive an analytic expression for the posterior of $\lambda$ as a function of the observed number of steps $s$. You can assume a flat (uniform) prior for $\lambda$.

### Answer:

* First of all, since we are talking about Posteriors and Priors, we can right away see that we need to use the **Bayes Theorem!**

$$ P(Z | X) = \frac{P(X | Z) \cdot P(Z)}{P(X)} $$

* **In here:**
    * $P(Z | X)$: Posterior -> Probability of **Hypothesis $Z$** given the **Observation $X$**!
    * $P(X | Z)$: Likelihood -> Probability of **Observation $X$** given the **Hypothesis $Z$**!
    * $P(Z)$: Prior -> Our Prior Estimate of the Truth of **Hypothesis $Z$**!
    * $P(X)$: Evidence -> Probability of **Observation/Data $X$** marginalized over all **Hypotheses $Z$**!

* For our case, we can write the Bayes' Theorem as follows:

$$ P(\lambda | s) = \frac{P(s | \lambda) \cdot P(\lambda)}{P(s)} $$

* **In here:**
    * $P(\lambda | s)$: Posterior -> Probability of **Hypothesis $\lambda$** given the **Observation $s$**!
    * $P(s | \lambda)$: Likelihood -> Probability of **Observation $s$** given the **Hypothesis $\lambda$**!
    * $P(\lambda)$: Prior -> Our Prior Estimate of the Truth of **Hypothesis $\lambda$**!
    * $P(s)$: Evidence -> Probability of **Observation/Data $s$** marginalized over all **Hypotheses $\lambda$**! (Marginal Likelihood)

* From the question we know that:
    * 1-) **Our Prior is Flat (Uniform) for $\lambda$!**
    * 2-) **$s$ follows a Poisson Distribution!**

* From here we can write the following expression for Likelihood, since $s$ follows Poisson Distribution!

$$ P(s | \lambda) = \frac{e^{-\lambda} \lambda^s}{s!} $$

* From here we can also write the Prior as follows, since $\lambda$ follows Uniform Distribution! (**$k$ is a constant!**)

$$ P(\lambda) = k $$

<h5><center>(In Uniform Distributions the distribution is one value only which we defined as $k$ here!)</center></h5>

* From here we can finally also write the Evidence as follows.

$$ P(s) = \int_{0}^{\infty} P(s | \lambda) \cdot P(\lambda) d\lambda $$

$$ P(s) = \int_{0}^{\infty} \frac{e^{-\lambda} \lambda^s}{s!} \cdot k d\lambda $$

<h5><center>(Because we marginalize over all Hypotheses ($\lambda$) we take the integral!)</center></h5>

* Using these new expressions, we can write the Bayes' Theorem as follows:

$$ P(\lambda | s) = \frac{\frac{e^{-\lambda} \lambda^s}{s!} \cdot k}{\int_{0}^{\infty} \frac{e^{-\lambda} \lambda^s}{s!} \cdot k d\lambda} $$

* From here:

$$ P(\lambda | s) = \frac{e^{-\lambda} \lambda^s \cdot \frac{k}{s!}}{\int_{0}^{\infty} e^{-\lambda} \lambda^s \cdot \frac{k}{s!} d\lambda} $$

$$ P(\lambda | s) = \frac{e^{-\lambda} \lambda^s \cdot \frac{k}{s!}}{ \frac{k}{s!} \cdot \int_{0}^{\infty} e^{-\lambda} \lambda^s d\lambda} $$

$$ P(\lambda | s) = \frac{e^{-\lambda} \lambda^s}{\int_{0}^{\infty} e^{-\lambda} \lambda^s d\lambda} $$

* Since the integral on the denominator represents a **"Normalization Constant"** we can write the following expression for $P(\lambda | s)$ i.e. Posterior:

$$ P(\lambda | s) \propto e^{-\lambda} \lambda^s $$
