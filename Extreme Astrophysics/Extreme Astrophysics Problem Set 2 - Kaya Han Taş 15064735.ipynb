* First, we define the Specific Intensity.

```python
#We import matplotlib for plotting purposes.
import matplotlib.pyplot as plt

#We also import numpy for our mathematical calculations.
import numpy as np

def Specific_Intensity(theta, tau, absorption_coeff, emission_coeff):
    """Takes the Theta/Angles (in degrees), Optical Depth (unitless), Absorption 
    Coefficient (in 1/cm) and Emission coefficient (in erg/cm^3 sr).
    
    Returns the Specific Intensity values for different angles/mu values."""
    #First we find the Source Function
    source_function = emission_coeff / absorption_coeff
    
    #Now we can find the Specific Intensity for different angles.
    #We convert the angles from degrees to radians by np.deg2rad!
    specific_intensity = source_function * (1 - np.exp(-tau*np.cos(np.deg2rad(theta))))
    
    return specific_intensity
```

* Now we plot Specific Intensity for different values of Optical Depth.

```python
#Now we define a range for our angles from 0 to 90. (Due to axial symmetry)
#Note that in our function they will be turned to radians since the numpy
#cosine function is defined for radians.
#So our interval will actually be 0 to œÄ/2.
angles = np.arange(0, 90, 0.1)

#Now let's plot the angular dependence of Specific Intensity.
plt.figure() #Creating the figure

#We plot our Specific Intensities for different Optical Depths.
#For œÑ = 0.1
plt.plot(angles, Specific_Intensity(angles, 0.1, 1e2, 1e12), color="red", label = "œÑ = 0.1")

#For œÑ = 1
plt.plot(angles, Specific_Intensity(angles, 1, 1e2, 1e12), color="orange", label = "œÑ = 1")

#For œÑ = 10
plt.plot(angles, Specific_Intensity(angles, 10, 1e2, 1e12), color="blue", label = "œÑ = 10")

#We also add Labels and Title.
plt.title("Angular Dependence of Specific Intensity for different Optical Depths (œÑ)")
plt.xlabel("Angles [deg]")
plt.ylabel("Intensity [erg]/[cm^2][sr]")

#We also add a Legend to show our labels for different temperatures.
plt.legend(loc = "upper right")

plt.show()
```
