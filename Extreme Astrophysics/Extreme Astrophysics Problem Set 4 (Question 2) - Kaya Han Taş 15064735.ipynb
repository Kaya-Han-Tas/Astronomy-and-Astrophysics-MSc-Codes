* Use your derived expression to find the Bremsstrahlung Spectrum $dW/d\omega$ from the single electronâ€“ion system, which should only be in terms of some constants and $b$, $v$ and $\omega$. 
* Plot this function with labelled axes showing the spectrum as a function of $\omega$.

```python
#Importing the libraries we need.
import numpy as np
import matplotlib.pyplot as plt
import math

#We first define all the Constant values we have in cgs units!

Z = 1 #Hydrogen Ion Charge
e = 5e-10 #Charge of Electron
b = 1e-5 #Closest approach to the ion in cm
c = 3e10 #Speed of Light
velocity = 0.8 * c #Velocity of the electron is close to speed of light
mass_e = 1e-27 #Mass of the Electron
```

```python
#Now we define the function for the Power of the Bremsstrahlung.
def Bremsstrahlung(omega_interval):
    """Takes the frequency interval, returns the Power of the Bremsstrahlung
    for each frequency given."""
    
    power1 = (2*np.pi)/3
    power2 = (Z**2) * (e**6) * np.exp(-(2*b*omega_interval)/velocity)
    power3 = (mass_e**2) * (b**2) * (velocity**2) * (c**3)
    
    power = power1 * (power2/power3)
    
    return power
```

```python
#Now we define an Ï‰ interval.
#Since the maximum Ï‰ is v/b we can start from zero to there.
omega_interval = np.arange(0, velocity/b, 1e10)

#Now we get our Graph of Intensity vs Time
plt.figure() #Creating the figure

#We plot the Power with the frequency!
plt.plot(omega_interval, Bremsstrahlung(omega_interval))

#We also add Labels and Title.
plt.title("The Change in Bremsstrahlung Power per Frequency")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Power [erg/sec]")
plt.xscale("log")
plt.yscale("log")

plt.show()

#The shape is expected from a Bremsstrahlung Spectrum since it happens very fast
#and electron doesn't move much so it doesn't change the frequency much.
#The cut-off happens when the electron stops feeling the ion!
```
